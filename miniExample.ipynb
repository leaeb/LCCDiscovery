# mini example: lcc discovery

this notebook shows a minimal end-to-end run of the lcc discovery algorithm on small synthetic data.

```python
import os
import sys

import numpy as np

#make sure the local module in this folder is importable
sys.path.insert(0, os.getcwd())

from LCCdiscovery import LCCConfig, discover_lcc
```

```python
rng = np.random.default_rng(7)

n = 60
m = 4
planted = np.array([0, 1, 2, 3, 4, 5])

#build m symmetric distance matrices from random points
dists = []
for _ in range(m):
    x = rng.normal(size=(n, 2))
    diff = x[:, None, :] - x[None, :, :]
    d = np.sqrt(np.sum(diff * diff, axis=-1))
    dists.append(d)

#make the planted set consistently close across modalities
for k in range(m):
    dk = dists[k]
    for i in planted:
        for j in planted:
            if i != j:
                dk[i, j] *= 0.03

cfg = LCCConfig(alpha=0.05, method="exact", fdr="bh", min_component_size=3)
res = discover_lcc(dists, cfg)

print("n components:", len(res.components))
print("component sizes:", [len(c) for c in res.components])
print("n noise:", len(res.noise))
print("n mutual edges:", int(res.mutual_adjacency.sum() // 2))
```
